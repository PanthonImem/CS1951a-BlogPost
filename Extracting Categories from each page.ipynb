```
import pandas as pd
import numpy
from bs4 import BeautifulSoup, Comment
import requests
import sqlite3
import json
import os
import urllib
import re
```

```
with open('names.json', 'r') as f:
    names = json.load(f)

S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

categories = {}

for counter, e in enumerate(names):
    title = e[0]

    PARAMS = {
      "action": "query",
      "format": "json",
      "titles": title,
      "prop": "categories"
    }

    response = S.get(url=URL, params=PARAMS).json()
    page = next(iter(response['query']['pages'].values()))
    L = []
    
    try:
        for category in page['categories']:
            L.append(category['title'])
    except:
        pass
    categories[title] = L

    if((counter+1)%5000 == 0):
        n = str(int((counter+1)/5000))
        with open('categories' + n + '.json', 'w') as fp:
            json.dump(categories, fp)

with open('categories_final.json', 'w') as final_f:
    json.dump(categories, fp)
```
